```python
essay_cols = ["essay0", "essay1", "essay2", "essay3", "essay4", "essay5", "essay6", "essay7", "essay8", "essay9"]
all_essays = dating[essay_cols].replace(np.nan, "", regex = True) #Must save this change to "NaN" values to a new variable (dataframe).
#all_essays.head()
all_essays = all_essays[essay_cols].apply(lambda x: " ".join(x), axis = 1)
#all_essays.head()
dating["essay_length"] = all_essays.apply(lambda x: len(x))

essays_answered = []
for item in essay_cols:
    new_item = dating[item].notnull() #https://stackoverflow.com/questions/29530232/how-to-check-if-any-value-is-nan-in-a-pandas-dataframe
    essays_answered.append(new_item)
essay_df = pd.concat(essays_answered, axis = 1)
dating["number_of_essays_answered"] = essay_df.sum(axis=1) #"axis = 1" specifies that I want to sum across columns, NOT rows.
#print(essay_df[0:30])

dating["avg_essay_length"] = dating.essay_length / dating.number_of_essays_answered

dating["me_frequency"] = all_essays.apply(lambda x: x.count("me"))
dating["i_frequency"] = all_essays.apply(lambda x: x.count("i"))

#Resources for summing: https://stackoverflow.com/questions/24810526/countif-in-pandas-python-over-multiple-columns-with-multiple-conditions/24811478
#  https://stackoverflow.com/questions/25748683/pandas-sum-dataframe-rows-for-given-columns

#Python lambda: https://www.w3schools.com/python/python_lambda.asp

#Sum in a dataframe: https://stackoverflow.com/questions/25748683/pandas-sum-dataframe-rows-for-given-columns

#Conditions: df[df['column_we_know'] == 'value_we_know']['column_we_want']
#  businesses[businesses.business_id == "5EvUIR4IzCWUOm0PsUZXjA"]["stars"]
#  https://datatofish.com/if-condition-in-pandas-dataframe/

dating.head()
```

```python
#dating.drinks.value_counts()
drinking = {"desperately": 5, "very often": 4, "often": 3, "socially": 2, "socially": 1, "not at all": 0}
dating["drinks_category"] = dating.drinks.map(drinking)
dating["drinks_category"] = dating["drinks_category"].replace(np.nan, 0, regex = True)
print(dating[["drinks", "drinks_category"]].head())
```

```python
#dating.smokes.value_counts()
smoking = {"yes": 4, "sometimes": 3, "when drinking": 2, "trying to quit": 1, "no": 0}
dating["smokes_category"] = dating.smokes.map(smoking)
dating["smokes_category"] = dating["smokes_category"].replace(np.nan, 0, regex = True)
print(dating[["smokes", "smokes_category"]].head())
print(dating["smokes_category"].value_counts())
print(dating["smokes_category"].unique())
```

```python
#dating.drugs.value_counts()
takes_drugs = {"often": 2, "sometimes": 1, "never": 0}
dating["drugs_category"] = dating.drugs.map(takes_drugs)
dating["drugs_category"] = dating["drugs_category"].replace(np.nan, 0, regex = True)
print(dating[["drugs", "drugs_category"]].head())
```

```python
#Normalization, classification, regression techniques.  Accuracy, precision, recall...

#KNN:
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split

key_data = dating[["drugs_category", "smokes_category", "drinks_category", "essay_length", "number_of_essays_answered", "age"]]
#key_data.isnull().any() #test for any "NaN" values in the data.
#print(key_data.head())
scaled_data = scale(key_data)
#print(scaled_data[0])


data_labels = dating.sign_category

train_data, test_data, train_labels, test_labels = train_test_split(scaled_data, data_labels, test_size = 0.2, random_state = 1) 

classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(train_data, train_labels)
print("Based on the score where k=5, the independent variables in the model predict the dependent variable (sign) {} of the time.".format(classifier.score(test_data, test_labels)))

scores = []
for k in range(1,101):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(train_data, train_labels)
    scores.append(classifier.score(test_data, test_labels))
plt.plot(range(1,101), scores)
plt.xlabel("k")
plt.ylabel("Validation Accuracy")
plt.show()
```

```python
#Naive Bayes
#Analyze all essay text to predict education level
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

unique_income = dating.income.unique()

income_category_dict = {}
for income in range(len(unique_income)):
    income_category_dict[unique_income[income]] = income
print(income_category_dict)
dating["income_category"] = dating.income.map(income_category_dict)
dating["income_category"] = dating["income_category"].replace(np.nan, 0, regex = True)
#print(dating["income_category"].head())

dating["all_essays"] = all_essays
#print(dating.all_essays.head())

train_data, test_data, train_labels, test_labels = train_test_split(dating["all_essays"], dating["income_category"], test_size = 0.2, random_state = 1)

counter = CountVectorizer()
counter.fit(train_data)
#print(counter.vocabulary_)

train_counts = counter.transform(train_data)
test_counts = counter.transform(test_data)


classifier = MultinomialNB()
classifier.fit(train_counts, train_labels)
predictions = classifier.predict(test_counts)
print(predictions)

print(accuracy_score(test_labels, predictions)) #this is the percentage of essays that were categorized correctly into income level.
```

```python
#Multiple Linear Regression
# How does essay essay length, number of essays answered, and age predict income level?
from sklearn.linear_model import LinearRegression

data_subset = dating[["essay_length", "number_of_essays_answered", "age"]]

x_train, x_test, y_train, y_test = train_test_split(data_subset, dating["income_category"], test_size = 0.2, random_state = 1)

model = LinearRegression()
model.fit(x_train, y_train)
print("R^2 is the coefficient of determination, or a measure of how much of the variance in our dependent variable, the predicted income level, is explained by our independent variables, our feature data of essay length, # of essays answered, and age.")
print("R^s is : {}".format(model.score(x_test, y_test)))
#model.coef_
#sorted(list(zip(['average_review_length','average_review_age'],model.coef_)),key = lambda x: abs(x[1]),reverse=True)

#Try this for education level as well...
```
